**1. Rondo Form is a type of musical structure, in which there is a
recurring theme/refrain, notated as A. Here are the rules for valid

-   Rondo forms always start and end with an A section.

-   In between the A sections, there should be contrasting sections

-   There shouldn't be any repeats in the sequence (such as ABBACCA).

Create a function which validates whether a given string is a valid
Rondo Form.

**Examples:**  
valid_rondo("ABACADAEAFAGAHAIAJA") ➞ True  
valid_rondo("ABA") ➞ True  
valid_rondo("ABBACCA") ➞ False

valid_rondo("A") ➞ False
